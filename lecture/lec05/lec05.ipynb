## Quick Check 3

```python
for counter in np.arange(5):
    print(9 * (counter + 1))
```

```python
multiplier = 1
while multiplier <= 5:
    print(9 * multiplier)
    multiplier += 1
```

<br/><br/>

```python
def sum_up_squares(start, end):
    counter = start
    total = 0
    while counter <= end:
        total += counter ** 2
        counter += 1
    return total

sum_up_squares(1, 5)
```
