## Conditional Statements

`if` clause

```python
temperature = 20
```

```python
clothing = "shirt"

if temperature < 32:
    clothing = "jacket"

print(clothing)
```

`elif` clauses

```python
clothing = "shirt" 
if temperature < 0:
    clothing = "snowsuit"
elif temperature < 32:
    clothing = "jacket"

print(clothing)
```

`else` clause

```python
if temperature < 0:
    clothing = "snowsuit"
elif temperature < 32:
    clothing = "jacket"
else:
    clothing = "shirt" 

print(clothing)
```

Conditional Statements in Functions

1. 

```python
def get_number_sign1(num):
    if num < 0:
        sign = "negative"
    elif num > 0:
        sign = "positive"
    else:
        sign = "neutral" 
    return sign
```

```python
get_number_sign1(50)
```
